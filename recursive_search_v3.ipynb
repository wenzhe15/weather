```python
import os
import json
import numpy as np
import pandas as pd

target = '.' # recursive search 를 시작할 디렉토리 위치

def get_json(target):
    geojson_full_path = []
    for root, dirnames, files in os.walk(target): # search를 시작하는 위치부터 모든 디렉토리를 돌면서 
        for x in files:
            if x.endswith('.geojson'):            # 확장자가 geojson 파일만 조건으로 찾아옵니다.
                full_path = root +'\\' + x
                geojson_full_path.append(full_path) # 전체경로를 geojson_full_path 에 담아서 반환합니다.
    return geojson_full_path

print(get_json(target))
```

```python
def make_df(features):
    geoshape = [np.array(f['geometry']['coordinates']).shape for f in features] # 좌표값을 numpy를 활용해 shape로 변환합니다.
    geotype = [f['geometry']['type'] for f in features]
    #geoname = [i['properties'] for i in features]
    geoname = [i['properties']['name'] for i in features]
    gtype = [i['type'] for i in features]
    df = pd.DataFrame({'gtype':gtype,
                       'geo_type':geotype, 
                       'geo_shape':geoshape, 
                       'geoname':geoname}) # 생성된 리스트를 활용해 데이터프레임을 만들어 반환합니다.
    return df
    
```

```python
target = '.'
for filename in get_json(target):  # 받아온 경로를 하나씩 불러와 json 파일을 읽습니다.
    with open(filename, 'r') as f:
        json_data = json.load(f)
        print(filename)
        if json_data['type'] == "FeatureCollection": 
            features = json_data["features"]
            print(len(features))
            df = make_df(features)
            print(df)
        else:
            temp = []
            temp.append(json_data)
            features = temp
            df = make_df(features)
            print(df)  
        print()
```

### 엑셀 시트에 담아서 확인
